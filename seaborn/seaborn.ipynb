# Visualization for Categorical Variable
## Box Plot

A box plot describes the distribution of a continuous variable by plotting the summary of 
five statistical terms: minimum, maximum, second quartile (50th percentile), first quartile
(25th percentile) and third quartile (75th percentile) in the data set. 

Box plots are a measure of how well distributed is the data in a data set. 

It divides the data set into three quartiles. It is possible to draw multiple boxes 
considering a categorical variable with different values. It is also useful in comparing 
the distribution of data across data sets by drawing box plots for each of them. 


```python
#Creating a box plot for "fare" with respect to "class".
sns.boxplot (x=titanic['class'],y=titanic['fare'], 
             data=titanic).set_title('Box plot for fare with respect to class')
```

## Violin Plot

A Violin plot is a combination of box plot and kernel density plot. So, these plots are 
easier to analyse and understand the distribution of the data. Similar to box plot, a 
violin plot can produce multiple violins corresponding to different values of a 
categorical variable. The quartile and whisker values from the boxplot are shown inside 
the violin. As the violin plot uses KDE, the wider portion of violin indicates the higher 
density and narrow region represents relatively lower density. The Inter-Quartile range 
in boxplot and higher density portion in “kde” fall in the same region of each category 
of violin plot.


```python
#Creating a violin plot for age with respect to sex.
sns.violinplot (x="class", y="age", hue="sex",
                data=titanic).set_title('Violin plot for showing Age with class and sex')
```

## Point Plot

The point plot is created using the pointplot() function and shows the point with respect 
to each vale of the categorical variable and joins them with a line segment.


```python
#Creating a point plot for "age" and "class".
sns.pointplot(y=titanic ['age'], x=titanic ['class'], 
              data=titanic).set_title('Point plot for age with respect to class')
```

# Line Plot

It is a plot that connects a series of points by drawing line segments between them. 
These points are ordered in one of their coordinate (usually the x-coordinate) value. 
Line charts are usually used in identifying the trends in data.


```python
#Creating a line plot for "fare" and "embarked".
sns.lineplot(y=titanic['fare'],x=titanic ['embarked'], 
             data=titanic).set_title('Line plot for fare and embarked')
```

# Count plot

This plot is drawn using countplot() and is used to show the number of observations 
corresponding to the different values of column specified in the function.


```python
#Creating a count plot for "alive".
sns.countplot (y=titanic ['alive'], data=titanic).set_title('Count plot for alive')
```

# Bar Plot

The bar plot draws a bar to display the relationship between a continuous and categorical 
variable.


```python
#Creating a bar plot for fare with respect to embarked.
sns.barplot (y=titanic['fare'],x=titanic ['embarked'], 
             data=titanic).set_title('Bar plot for fare with respect to embarked')
```

# Strip Plot

The function stripplot() is also used when one of the variable under study is categorical.
It represents the data in sorted order along any one of the axis. It is similar to scatter
plot which is created with two continuous variables. The major problem with the plot is 
that the points are overlapped. To handle this situation, the argument ‘Jitter’ is 
included in stripplot()  with Boolean value as True to add some random noise to the data. 
This parameter will adjust the positions along the categorical axis.


```python
#Strip plot with jitter argument for fare with class.
sns.stripplot(x = "class", y = "fare", jitter =True, 
                  data =titanic).set_title('Strip Plot for class and fare')

```

# Swarm Plot

The utility of swarmplot() is same as another option which can be used as an alternate to 
‘Jitter’ discussed in strip plot. This function positions each point of scatter plot on 
the categorical axis and thereby avoids overlapping points.


```python
sns.swarmplot(x = "class", y = "fare", 
                  data =titanic).set_title('Swarm Plot for class and fare')
```

# Facet Grid

This chart helps in visualizing distribution of one variable and relationship between multiple variables separately within subsets of dataset using multiple panels. A FacetGrid can be drawn using three dimensions − row, col, and hue. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.



```python
import matplotlib.pyplot as plt
#Creating a facet grid.
sns.FacetGrid (titanic, col = "class")
#Creating facet grid of histogram for "age" and "class". 
gridl = sns.FacetGrid (titanic, col = "class")
gridl.map (plt.hist, "age")
#Creating facet grid of histogram for "age","class", "alive". 
grid2=sns.FacetGrid (titanic, col = "class", hue="alive") 
grid2.map(plt.hist, "age")
#Creating facet grid of scatter chart for "age", "sex", "fare". 
grid3=sns.FacetGrid (titanic, col = "class", hue="sex") 
grid3.map (plt.scatter, "age", "fare")
```

## Scatter Plot

A scatter plot using “seaborn” is drawn through scatterplot().


```python
iris = sns.load_dataset ('iris')
print ("The dimension of the dataset is: ", iris.shape)
print ("The columns in the dataset are: \n", iris.keys())
print ("The details of the titanic dataset is: \n", iris.describe())
```

```python
#Scatter plot for "sepal_length" and "petal_length".
sns.scatterplot (y="sepal_length",x="petal_length", 
                 data=iris).set_title('Scatter plot for sepal_length and petal_length')
```

## Regression Plot

There are two main functions in Seaborn to visualize a linear relationship determined 
through regression: regplot() and lmplot(). These regression plot helps to draw a line of 
best fit on the scatterplot. 


```python
#Using regplot() for "petal_length" and "petal_width". 
sns.regplot (y="petal_width", x="petal_length", data=iris)
#Using Implot() for "sepal_width" and "petal_width". 
sns.lmplot (y="sepal_width", x="petal_width", data=iris)
sns.lmplot (y="sepal_width", x="petal_width", data=iris,order=2)
```

## Univariate Distribution Plot

The function distplot() helps us to visualize the parametric distribution of a dataset. 
The two most important arguments are kde (Kernel Density Plot) and hist (Histogram) which
has a Boolean value. If the value are true, the plot shows both the charts. 


```python
plt.figure(figsize=(10,10))
#Univariate distribution Plot for "petal_length". 
plt.subplot(311)
sns.distplot(iris[['petal_length']]).set_title('Distribution Plot for petal_length')
plt.subplot(312)
sns.distplot(iris[['petal_length']],
             kde=False).set_title('Distribution Plot for petal_length for kde False')
plt.subplot(313)
sns.distplot(iris[['petal_length']],
             hist=False).set_title('Distribution Plot for petal_length for hist False')
```

## Joint Plot

Joint plot is used for bivariate Distribution is used to determine the relation between 
    two variables. This mainly deals with relationship between two variables and how one 
    variable is behaving with respect to the other. The jointplot() function creates a 
    multi-panel figure that projects the bivariate relationship between two variables and 
    also the univariate distribution of each variable on separate axes.


```python
#Joint plot for "sepal.length" and "petal_length".
sns.jointplot (y="petal_length", x="sepal_length", data=iris)
#Regression Plot for "sepal.length" and "petal_length".
sns.jointplot (y="petal_length", x="sepal_length", kind="reg", data=iris)
```

## Joint Hexbin Plot

It is possible to create a hexbin plot by considering hex as the value of kind in the 
jointplot(). Hexagonal binning is used in bivariate data analysis when the data is sparse 
in density i.e., when the data is very scattered and difficult to analyze through 
scatterplots.


```python
sns.jointplot(y="petal_length", x="sepal_length", kind="hex", data=iris)
```

## Joint Kernel Density Plot

For creating a kernel density estimation plot, we need to add kind argument to the 
jointplot() with value=”kde”.


```python
sns.jointplot(y="petal_length", x="sepal_length", kind="kde", data=iris)
```

## Pairplot 

The pairplot()  function produces a scatterplot between all possible pairs of variables 
in a dataset and for each variable, it uses the same scale. It is also possible to create
a regression line on all the plots using the “reg” value for kind argument. It is also 
possible to draw a pair plot for selected columns.


```python
sns.pairplot(data=iris,hue='species')
```

```python
sns.pairplot(data=iris,hue='species',kind='reg')
```

## Pair Grid

This plot helps us define the type of chart on the diagonal plot along with specifications
for the type of chart on upper , lower and off diagonal separately. Unlike FacetGrid, it
uses different pair of variable for each subplot. It forms a matrix of sub-plots. The 
usage of pairgrid is similar to facetgrid. We need to first initialise the grid and then 
pass the type of chart inside the appropriate function. 


```python
#Creating the pair grid with scatter plot. 
gridl=sns.PairGrid (iris)
gridl.map(plt.scatter)
```

```python
#Creating the pair grid with scatter plot.
grid2 = sns.PairGrid (iris)
grid2.map_upper(plt.scatter) 
grid2.map_lower(sns.kdeplot) 
grid2.map_diag (plt.hist)
```
